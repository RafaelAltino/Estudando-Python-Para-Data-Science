```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris, hue='species')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
h = sns.FacetGrid(tips, col='time', row='sex')
h.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm', markers=['o', 'v'])
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='sex', row='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='day', aspect=1.2, height=8)
```

```python
import seaborn as sns
import numpy as np
%matplotlib inline
sns.set(style='darkgrid')
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#Coluna com a média dos sexos, linha com o desvio padrão da média
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
#estimator caso queira que faça outro calculo nos dados. Como desvio padrão
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data= tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(data=tips, orient='h')
```

```python
sns.violinplot(x='day', y='tip', data=tips)
```

```python
sns.violinplot(x='sex', y='tip', data=tips, hue='day')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm')
```

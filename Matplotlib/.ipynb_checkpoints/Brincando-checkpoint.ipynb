```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pkm = pd.read_excel('Tabela Pokemon.xlsx')
```

```python
pkm
```

```python
qnt = ((pkm['Tipo 1'].value_counts()) + (pkm['Tipo 2'].value_counts()))
qnt.sort_index(inplace = True)

tipos = np.sort(pkm['Tipo 1'].unique())
```

```python
fig = plt.figure(figsize=(12,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(tipos, qnt)
```

```python
tipos = np.sort(pkm['Tipo 1'].unique())
tipos
```

```python
fig, ax = plt.subplots(figsize=(15,4))
ax.bar(tipos, qnt)
ax.set_xlabel('Tipo')
ax.set_ylabel('Quantidade')
ax.set_title('Quantidade de Pokemons por tipo')
```

```python
unicos = pkm[pd.isna(pkm['Tipo 2'])]['Tipo 1'].value_counts()
unicos.sort_index(inplace = True)
unicos
```

```python
fig, ax2 = plt.subplots(figsize=(15,4))
ax2.bar(tipos, unicos)
ax2.set_xlabel('Tipo')
ax2.set_ylabel('Quantidade')
ax2.set_title('Quantidade de Pokemons de tipo Ãºnico')
```

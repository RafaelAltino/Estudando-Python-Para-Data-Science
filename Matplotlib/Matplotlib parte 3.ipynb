```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x * x
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
#color para cor, pode ser nome em ingles ou hexadecimal
#linewidht ou só lw para grossura da linha
#alpha para transparência da linha, útil quando se tem várias linhas
#linestyle para estili da linha
ax.plot(x, x**2, color='#8B008B', linewidth=10, alpha=0.5, linestyle=':')
ax.plot(x, x**3)

#Define os valores mínimo e máximo do eixo x exibidos no gráfico
ax.set_xlim([0, 2])
#Define os valores mínimo e máximo do eixo y exibidos no gráfico
ax.set_ylim([0,10])
```

```python
#Apenas plotar os pontos onde há dados
plt.scatter(x, y)
```

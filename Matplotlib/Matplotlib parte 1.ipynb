```python
#Importação padrão do matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Permite que os gráficos sejam exibidos nas linhas logo após fazer o plot deles
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
```

```python
y = x * x
```

```python
#Cria um gráfico passando seu (eixo x, eixo y, cor/estampa)
plt.plot(x, y, 'ro')

#Indica o título de cada eixo
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
#Indica o título do gráfico
plt.title('Título')
```

```python
#Cria uma matriz de gráficos, passando(linhas, colunas, qual irei definir a seguir)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
#Agora especifico que irei definir o segundo gráfico
plt.subplot(1, 2, 2)
plt.plot(x, y, 'g-')
```

```python
#Crio gráficos com maior controle em suas definições
#Crio uma figura de um gráfico, como um 'molde'
fig = plt.figure()
#Defino as especificações do tamanho do primeiro gráfico passando uma lista que contenha:
#[espaço à esquerda, espaço abaixo, largura, altura] sendo 1.0 = 100% e 0.2 = 20%
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
#Passo seus dados e cores
axes1.plot(x, y)
#Chama a função de setar título do eixo x
axes1.set_xlabel('Eixo x')
#Chama a função de setar título do gráfico
axes1.set_title('Título')

#Especifico o tamanho do segundo gráfico
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])
#Passo seus valores e cores
axes2.plot(y, x, 'r')
```

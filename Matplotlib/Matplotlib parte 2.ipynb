```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
```

```python
y = x * x
```

```python
#Cria uma figura e um eixo
fig, ax = plt.subplots()
#Plota as linhas dos eixos
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'g-.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('TÃ­tulo')
```

```python
#Cria figura e eixos, indicando numero de linhas e colunas da figura
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
#acessa cada grÃ¡fico atravÃ©z de iteraÃ§Ã£o
for axe in ax:
    axe.plot(x, y, 'b')
```

```python
fig
```

```python
#Acessa atravÃ©s de indexaÃ§Ã£o. ax[x][y] para matrizes
ax[0].plot(x, y, 'r')
ax[1].plot(-2*x, 3*y, 'g')
fig
```

```python
#Cria figura e eixo grande
fig, ax = plt.subplots(nrows=5, ncols=5)
#Cria espaÃ§amento entre cada grÃ¡fico
plt.tight_layout()
```

```python
#Cria uma figura e especifica o tamanho, para depois adicionar os eixos
fig = plt.figure(figsize=(7, 3), dpi=100)
#Adicionando eixo
ax = fig.add_axes([0.2, 0.2, 0.5, 0.8])
ax.plot(x, y)
```

```python
#JÃ¡ cria figura e eixo passando o tamanho por parÃ¢metro
fig, axes = plt.subplots(figsize=(12, 6))
#Define o nome da linha
axes.plot(x, y, 'r', label='x ^ 2')
axes.plot(y, x, 'g--', label = 'x ^ 1/2')
#Exibe os nomes das linhas como legenda
axes.legend()
axes.set_title('TÃ­tulo')
```

```python
#Salva a figura como uma imagem no diretÃ³rio do notebook
fig.savefig('figura.png')
```

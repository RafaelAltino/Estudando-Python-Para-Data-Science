```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2,
                  centers=4, cluster_std=1.8, random_state=50)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(16,8))
ax1.set_title('Original')
ax1.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
ax2.set_title('KMeans')
ax2.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data[0])
f, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(16,8))
ax1.set_title('Original')
ax1.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
ax2.set_title('KMeans')
ax2.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')
```

```python
kmeans.cluster_centers_
```

```python
kmeans.inertia_
```

```python
distancia = []
for i in range(2, 12):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data[0])
    distancia.append(kmeans.inertia_)
```

```python
plt.plot(range(2,12), distancia)
```
